```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

## Data pre-processing

```python
df = pd.read_csv('pre-processed_data.csv')
df
```

```python
# Mapping target
def convert_sof_target(result: str) -> list:
    mapping = {'Dropout': 0, 'Enrolled': 1, 'Graduate': 2}
    
    if result in mapping:
        return mapping[result]
    
    raise ValueError('Not a valid label')

def convert_log_target(result : str) -> int:
    if (result == 'Dropout' or result == "Enroll"):
        return 0
    
    return 1
```

## Logistic regression

```python
df1 = df.copy()
df1['Target'] = df1['Target'].apply(convert_log_target)
data1 = df1.to_numpy()
X = data1[:, :-1]
Y = data1[:, -1]

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.2, random_state= 16)
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train,Y_train)
log_predicted = log_reg.predict(X_test)
acc = metrics.accuracy_score(Y_test, log_predicted)

print('Accuracy of logistic: ', acc)
```

# Softmax regression

```python
df2 = df.copy()
df2['Target'] = df2['Target'].apply(convert_sof_target)
data2 = df2.to_numpy()
X = data2[:, :-1]
Y = data2[:, -1]

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.2, random_state= 16)
```

```python
sof_reg = LogisticRegression(multi_class='multinomial')
sof_reg.fit(X_train,Y_train)
sof_predicted = log_reg.predict(X_test)
acc = metrics.accuracy_score(Y_test, sof_predicted)

print('Accuracy of softmax: ', acc)
```
